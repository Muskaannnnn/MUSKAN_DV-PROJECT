###HISTOGRAM

```python
climate=data['Summer']
plt.hist(climate,edgecolor="black",color="blue")
plt.xlabel("AVG TEMPERATURE")
plt.ylabel("NO OF COUNTRIES")
plt.title("Climatic conditions")
```

###SCATTER PLOT(Temp in Winter season of every country plotted)

```python
rain=data['Winter']
country=data['Country']
plt.figure(figsize=(15,5))
plt.scatter(country,rain,cmap="Blues",marker="x",label="Snowfall Between 1991 to 2020 ")
plt.xlabel("COUNTRY")
plt.ylabel("AVG TEMPERATURE")
plt.legend()
plt.show()
```

###PIE CHART(Importance of Avegrage temp )

```python
mean=[13.438000,4.172000,1.94500,5.118000,6.349000]
Designation=["ANNUAL","RAINY","SUMMER","WINTER","MONSOON"]
explodes = [0.1, 0, 0, 0, 0]
plt.pie(mean,labels=Designation,explode=explodes,startangle=90)
plt.show()
```

###LINE GRAPH(Plotted with Summer and Annual Temp)

```python
p=data['Annual']
summer=data['Summer']
fig,ax=plt.subplots()
ax.plot(p,summer,linestyle="-.",label="SUMMER")
ax.legend()
ax.set_title("COMPARSATION OF ANNUAL AND SUMMER TEMPERATURE")
ax.set_xlabel("AVG ANNUAL TEMPERATURE")
ax.set_ylabel("AVG SUMMER TEMPERATURE")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

import seaborn as sns

%matplotlib inline
```

###HEAT MAP(Shows the country with the highst heat signature that is hot temprature)

```python
fig, ax = plt.subplots(figsize=(8,8)) 
df2 = pd.pivot_table(values ='Winter', index ='Country', columns ='Annual', data = data)
sns.heatmap(df2,cmap='hot_r' , linecolor ='b', linewidths = 0.3,  ax=ax).set_title('Heat Map')
```

###PAIR PLOT(Every dataset is compared with each other)

```python
g=sns.pairplot(data, palette ='hot_r') 
g.fig.suptitle("Pair Plot")
g.add_legend()
```

###HIST PLOT(Here annual and monsoon temprature is Plotted over each other)

```python
annual=data['Annual']
monsoon=data['Monsoon']
sns.histplot(annual,color='blue',alpha=0.5)
sns.histplot(monsoon,color='red',alpha=0.6) 
```

Box Plot

```python
numeric=data.select_dtypes (include="number")
num_cols=numeric.columns
plt.figure(figsize=(30,15))
for i in range(len(num_cols)):
  plt.subplot (3,3,i+1) 
  sns.boxenplot (x=data[num_cols[i]]) 
  plt.title(num_cols[i])
plt.show()
```

```python
from PIL import Image 
import matplotlib.pyplot as plt 
from wordcloud import WordCloud, STOPWORDS
```

```python
#Stop Words
comment_words = '' 
stopwords = set(STOPWORDS) 

comment_words = data['Country'].str.lower()
comment_words = comment_words.to_string()

wc = WordCloud(width = 800, height = 800, background_color ='white', stopwords = stopwords, min_font_size = 10)
wc = wc.generate(comment_words)
```

```python
plt.figure(facecolor = None) 
plt.imshow(wc) 
plt.axis("off") 
plt.tight_layout(pad = -1) 
plt.show()
```
